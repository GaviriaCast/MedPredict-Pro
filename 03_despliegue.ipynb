## 2. Verificación del modelo guardado

````python
import joblib
import pandas as pd
import numpy as np
import os

archivos = ['models/best_rf_model.pkl', 'models/scaler.pkl', 'models/feature_names.pkl']
for f in archivos:
    existe = os.path.exists(f)
    print(f'  {f}: {"encontrado" if existe else "NO encontrado"}')

model         = joblib.load('models/best_rf_model.pkl')
scaler        = joblib.load('models/scaler.pkl')
feature_names = joblib.load('models/feature_names.pkl')

print(f'\nModelo: {type(model).__name__}')
print(f'Features ({len(feature_names)}): {feature_names}')
````

## 3. Prueba de predicción antes del despliegue

````python
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('data/chronic_disease_clean.csv')
le = LabelEncoder()
cat_cols = df.select_dtypes(include='object').columns
df_encoded = df.copy()
for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

TARGET = 'DISCAPACIDAD'
cols_excluir = ['SEDE', 'NOMBRE_DIAG', 'PLAN_BENEFICIOS', 'OCUPACION', 'TIPODISCAPAC',
                'NIVELEDUCATIVO', 'IDENTGEN', 'ORIGSEX', 'GRUPO ETNICO', TARGET]
X_sample = df_encoded.drop(columns=[c for c in cols_excluir if c in df_encoded.columns])
X_sample = X_sample.reindex(columns=feature_names, fill_value=0)

muestra        = X_sample.iloc[:5]
predicciones   = model.predict(muestra)
probabilidades = model.predict_proba(muestra)[:, 1]

print('Prueba con 5 registros del dataset:')
for i, (pred, prob) in enumerate(zip(predicciones, probabilidades)):
    label = 'Discapacidad' if pred == 1 else 'Sin discapacidad'
    print(f'  Registro {i+1}: {label} | Probabilidad={prob:.2%}')

print('\nModelo funcionando correctamente')
````

````python
with open('app.py', 'w', encoding='utf-8') as f:
    f.write('import streamlit as st\n')
    f.write('import joblib\n')
    f.write('import pandas as pd\n')
    f.write('import numpy as np\n')
    f.write('import time\n')
    f.write('\n')
    f.write('st.set_page_config(\n')
    f.write('    page_title="MedPredict Pro | Disability Risk",\n')
    f.write('    page_icon="🧠",\n')
    f.write('    layout="wide",\n')
    f.write('    initial_sidebar_state="collapsed"\n')
    f.write(')\n')
    f.write('\n')
    f.write('st.markdown("""\n')
    f.write('<style>\n')
    f.write('@import url(\'https://fonts.googleapis.com/css2?family=Inter:opsz,wght@14..32,300;400;600;700;800&display=swap\');\n')
    f.write('@import url(\'https://fonts.googleapis.com/css2?family=JetBrains+Mono:wght@300;400;500&display=swap\');\n')
    f.write('\n')
    f.write('* { margin:0; padding:0; box-sizing:border-box; }\n')
    f.write('\n')
    f.write('.stApp {\n')
    f.write('    background: radial-gradient(circle at 10% 20%, #0B0F19, #05080F);\n')
    f.write('    font-family: \'Inter\', sans-serif;\n')
    f.write('}\n')
    f.write('#MainMenu, footer, header { display: none; }\n')
    f.write('\n')
    f.write('[data-testid="stSidebar"] {\n')
    f.write('    background: rgba(10,15,25,0.75);\n')
    f.write('    backdrop-filter: blur(12px);\n')
    f.write('    border-right: 1px solid rgba(0,255,255,0.15);\n')
    f.write('}\n')
    f.write('\n')
    f.write('.hero {\n')
    f.write('    background: linear-gradient(120deg, rgba(0,224,255,0.08), rgba(124,58,237,0.08));\n')
    f.write('    border-radius: 32px;\n')
    f.write('    padding: 1.8rem 2rem;\n')
    f.write('    margin-bottom: 2rem;\n')
    f.write('    border: 1px solid rgba(0,224,255,0.25);\n')
    f.write('    overflow: hidden;\n')
    f.write('}\n')
    f.write('.hero h1 {\n')
    f.write('    font-size: 2.5rem;\n')
    f.write('    font-weight: 800;\n')
    f.write('    background: linear-gradient(135deg, #FFFFFF, #00e0ff);\n')
    f.write('    -webkit-background-clip: text;\n')
    f.write('    -webkit-text-fill-color: transparent;\n')
    f.write('}\n')
    f.write('.hero p { color: #94a3b8; margin-top: 0.5rem; }\n')
    f.write('\n')
    f.write('.metric {\n')
    f.write('    background: rgba(0,224,255,0.05);\n')
    f.write('    border-radius: 20px;\n')
    f.write('    padding: 1rem;\n')
    f.write('    text-align: center;\n')
    f.write('    border: 1px solid rgba(0,224,255,0.2);\n')
    f.write('}\n')
    f.write('.metric-value {\n')
    f.write('    font-size: 2rem;\n')
    f.write('    font-weight: 700;\n')
    f.write('    color: #00e0ff;\n')
    f.write('    font-family: \'JetBrains Mono\', monospace;\n')
    f.write('}\n')
    f.write('.metric-label {\n')
    f.write('    font-size: 0.7rem;\n')
    f.write('    text-transform: uppercase;\n')
    f.write('    letter-spacing: 0.08em;\n')
    f.write('    color: #94a3b8;\n')
    f.write('}\n')
    f.write('\n')
    f.write('.stButton > button {\n')
    f.write('    background: linear-gradient(90deg, #00e0ff, #7c3aed) !important;\n')
    f.write('    border: none !important;\n')
    f.write('    border-radius: 40px !important;\n')
    f.write('    padding: 0.7rem 1.5rem !important;\n')
    f.write('    font-weight: 600 !important;\n')
    f.write('    color: white !important;\n')
    f.write('    transition: transform 0.2s, box-shadow 0.2s !important;\n')
    f.write('}\n')
    f.write('.stButton > button:hover {\n')
    f.write('    transform: translateY(-2px) !important;\n')
    f.write('    box-shadow: 0 10px 25px -5px rgba(0,224,255,0.4) !important;\n')
    f.write('}\n')
    f.write('\n')
    f.write('.stSlider > div > div > div { background: #00e0ff !important; }\n')
    f.write('div[data-baseweb="select"] > div {\n')
    f.write('    background: #0f172a !important;\n')
    f.write('    border-color: #1e293b !important;\n')
    f.write('    border-radius: 12px !important;\n')
    f.write('    color: #e2e8f0 !important;\n')
    f.write('}\n')
    f.write('label[data-testid="stWidgetLabel"] {\n')
    f.write('    color: #94a3b8 !important;\n')
    f.write('    font-size: 0.72rem !important;\n')
    f.write('    text-transform: uppercase !important;\n')
    f.write('    letter-spacing: 0.05em !important;\n')
    f.write('}\n')
    f.write('\n')
    f.write('.stTabs [data-baseweb="tab-list"] {\n')
    f.write('    gap: 8px;\n')
    f.write('    background: rgba(30,41,59,0.4);\n')
    f.write('    border-radius: 40px;\n')
    f.write('    padding: 0.25rem;\n')
    f.write('}\n')
    f.write('.stTabs [data-baseweb="tab"] {\n')
    f.write('    border-radius: 32px;\n')
    f.write('    padding: 0.5rem 1.2rem;\n')
    f.write('    font-weight: 500;\n')
    f.write('    color: #94a3b8 !important;\n')
    f.write('}\n')
    f.write('.stTabs [aria-selected="true"] {\n')
    f.write('    background: linear-gradient(90deg, #00e0ff, #7c3aed) !important;\n')
    f.write('    color: white !important;\n')
    f.write('}\n')
    f.write('\n')
    f.write('.result-panel {\n')
    f.write('    background: rgba(15,25,40,0.7);\n')
    f.write('    backdrop-filter: blur(12px);\n')
    f.write('    border-radius: 32px;\n')
    f.write('    border: 1px solid rgba(0,224,255,0.3);\n')
    f.write('    padding: 2rem;\n')
    f.write('    margin-top: 2rem;\n')
    f.write('    position: relative;\n')
    f.write('    overflow: hidden;\n')
    f.write('}\n')
    f.write('.result-panel::before {\n')
    f.write('    content: \'\';\n')
    f.write('    position: absolute;\n')
    f.write('    top: 0; left: 0; right: 0;\n')
    f.write('    height: 2px;\n')
    f.write('    background: linear-gradient(90deg, #00e0ff, #7c3aed);\n')
    f.write('}\n')
    f.write('\n')
    f.write('.prob-bar-container {\n')
    f.write('    background: rgba(255,255,255,0.1);\n')
    f.write('    border-radius: 40px;\n')
    f.write('    height: 12px;\n')
    f.write('    overflow: hidden;\n')
    f.write('    margin: 0.5rem 0;\n')
    f.write('}\n')
    f.write('.prob-percent {\n')
    f.write('    font-family: \'JetBrains Mono\', monospace;\n')
    f.write('    font-size: 3rem;\n')
    f.write('    font-weight: 300;\n')
    f.write('    color: #00e0ff;\n')
    f.write('    line-height: 1;\n')
    f.write('}\n')
    f.write('.risk-badge {\n')
    f.write('    display: inline-block;\n')
    f.write('    padding: 0.4rem 1.2rem;\n')
    f.write('    border-radius: 40px;\n')
    f.write('    font-size: 0.85rem;\n')
    f.write('    font-weight: 600;\n')
    f.write('    margin-top: 0.5rem;\n')
    f.write('}\n')
    f.write('.risk-high { background: rgba(239,68,68,0.15); color: #ef4444; border: 1px solid rgba(239,68,68,0.4); }\n')
    f.write('.risk-med  { background: rgba(245,158,11,0.15); color: #f59e0b; border: 1px solid rgba(245,158,11,0.4); }\n')
    f.write('.risk-low  { background: rgba(16,185,129,0.15); color: #10b981; border: 1px solid rgba(16,185,129,0.4); }\n')
    f.write('\n')
    f.write('.footer {\n')
    f.write('    text-align: center;\n')
    f.write('    padding: 1.5rem;\n')
    f.write('    font-size: 0.68rem;\n')
    f.write('    color: #475569;\n')
    f.write('    border-top: 1px solid #1e293b;\n')
    f.write('    margin-top: 2rem;\n')
    f.write('    font-family: \'JetBrains Mono\', monospace;\n')
    f.write('    letter-spacing: 0.05em;\n')
    f.write('}\n')
    f.write('\n')
    f.write('p, .stMarkdown p { color: #e2e8f0 !important; }\n')
    f.write('h1,h2,h3,h4 { color: #e2e8f0 !important; }\n')
    f.write('.stExpander { border-color: #1e293b !important; }\n')
    f.write('</style>\n')
    f.write('""", unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('# ── Carga del modelo ──────────────────────────────────────────────────────────\n')
    f.write('@st.cache_resource\n')
    f.write('def load_models():\n')
    f.write('    model         = joblib.load(\'models/best_rf_model.pkl\')\n')
    f.write('    scaler        = joblib.load(\'models/scaler.pkl\')\n')
    f.write('    feature_names = joblib.load(\'models/feature_names.pkl\')\n')
    f.write('    return model, scaler, feature_names\n')
    f.write('\n')
    f.write('model, scaler, feature_names = load_models()\n')
    f.write('\n')
    f.write('# ── Encodings — solo features clínicas (16 features) ─────────────────────────\n')
    f.write('NORMAL_MAP     = {\'Normal\':2, \'No Registra\':1, \'Anormal\':0}\n')
    f.write('IMC_MAP        = {\'BAJO PESO\':0, \'DESNUTRICION LEVE\':1, \'DESNUTRICION MODERADA\':2,\n')
    f.write('                  \'DESNUTRICION SEVERA\':3, \'OBESIDAD I\':4, \'OBESIDAD II\':5,\n')
    f.write('                  \'OBESIDAD III\':6, \'PESO INSUFICIENTE\':7, \'PESO NORMAL\':8, \'SOBREPESO\':9}\n')
    f.write('DISNEA_MAP     = {\'LIGERA\':0, \'MODERADA\':1, \'MUY GRAVE\':2}\n')
    f.write('GENERO_MAP     = {\'Femenino\':0, \'Intersexual\':1, \'Masculino\':2, \'No Registra\':3}\n')
    f.write('CLASIFISUI_MAP = {\'Sin Riesgo Alta con Seguimiento Ambulatorio\':0,\n')
    f.write('                  \'Sin Riesgo Ambulatorio\':1, \'Riesgo Leve\':2,\n')
    f.write('                  \'Riesgo Moderado\':3, \'Riesgo Alto\':4, \'Riesgo Muy Alto\':5}\n')
    f.write('\n')
    f.write('# ── Hero ──────────────────────────────────────────────────────────────────────\n')
    f.write('st.markdown("""\n')
    f.write('<div class="hero">\n')
    f.write('    <div style="display:flex; justify-content:space-between; align-items:center; flex-wrap:wrap; gap:1rem;">\n')
    f.write('        <div>\n')
    f.write('            <div style="display:flex; gap:10px; margin-bottom:10px;">\n')
    f.write('                <span style="background:rgba(0,224,255,0.15);padding:0.2rem 0.8rem;border-radius:40px;font-size:0.7rem;color:#00e0ff;">⚡ CLINICAL AI</span>\n')
    f.write('                <span style="background:rgba(124,58,237,0.15);padding:0.2rem 0.8rem;border-radius:40px;font-size:0.7rem;color:#a78bfa;">🇨🇴 COLOMBIA 2026</span>\n')
    f.write('            </div>\n')
    f.write('            <h1>MedPredict Pro</h1>\n')
    f.write('            <p>Disability Risk Engine · Random Forest + GridSearchCV · 16 Features Clínicas</p>\n')
    f.write('        </div>\n')
    f.write('        <div style="display:flex; gap:1rem;">\n')
    f.write('            <div class="metric"><div class="metric-value">0.939</div><div class="metric-label">ROC-AUC</div></div>\n')
    f.write('            <div class="metric"><div class="metric-value">0.741</div><div class="metric-label">F1-Score</div></div>\n')
    f.write('            <div class="metric"><div class="metric-value">1,753</div><div class="metric-label">Pacientes</div></div>\n')
    f.write('            <div class="metric"><div class="metric-value">16</div><div class="metric-label">Features</div></div>\n')
    f.write('        </div>\n')
    f.write('    </div>\n')
    f.write('</div>\n')
    f.write('""", unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('# ── Layout principal ──────────────────────────────────────────────────────────\n')
    f.write('col_left, col_right = st.columns([2, 1], gap="large")\n')
    f.write('\n')
    f.write('with col_left:\n')
    f.write('    tab1, tab2, tab3 = st.tabs(["📋 Datos Clínicos", "🫀 Estado de Sistemas", "🦠 COVID-19"])\n')
    f.write('\n')
    f.write('    with tab1:\n')
    f.write('        col_a, col_b = st.columns(2)\n')
    f.write('        with col_a:\n')
    f.write('            edad = st.slider("Edad (años)", 0, 110, 60)\n')
    f.write('            peso = st.slider("Peso (kg)", 20, 175, 70)\n')
    f.write('        with col_b:\n')
    f.write('            talla   = st.slider("Talla (cm)", 100, 200, 160)\n')
    f.write('            imc_val = round(peso / ((talla/100)**2), 2)\n')
    f.write('            imc_cat = (\'BAJO PESO\' if imc_val < 18.5 else\n')
    f.write('                       \'NORMAL\'    if imc_val < 25   else\n')
    f.write('                       \'SOBREPESO\' if imc_val < 30   else \'OBESIDAD\')\n')
    f.write('            st.metric("IMC calculado", f"{imc_val} kg/m²", delta=imc_cat)\n')
    f.write('\n')
    f.write('        st.markdown("#### Escalas clínicas")\n')
    f.write('        col_c, col_d = st.columns(2)\n')
    f.write('        with col_c:\n')
    f.write('            resultado_imc = st.selectbox("Resultado IMC", list(IMC_MAP.keys()))\n')
    f.write('            escala_disnea = st.selectbox("Escala de disnea", list(DISNEA_MAP.keys()))\n')
    f.write('        with col_d:\n')
    f.write('            clasifisui = st.selectbox("Clasificación SUI", list(CLASIFISUI_MAP.keys()))\n')
    f.write('            genero     = st.selectbox("Género", list(GENERO_MAP.keys()))\n')
    f.write('\n')
    f.write('    with tab2:\n')
    f.write('        st.markdown("#### Estado de sistemas orgánicos")\n')
    f.write('        col_s1, col_s2 = st.columns(2)\n')
    f.write('        with col_s1:\n')
    f.write('            cardiovascular = st.selectbox("Cardiovascular", [\'Normal\',\'Anormal\',\'No Registra\'])\n')
    f.write('            pulmonar       = st.selectbox("Pulmonar",       [\'Normal\',\'Anormal\',\'No Registra\'])\n')
    f.write('            neurologico    = st.selectbox("Neurológico",    [\'Normal\',\'Anormal\',\'No Registra\'])\n')
    f.write('        with col_s2:\n')
    f.write('            mental        = st.selectbox("Mental",        [\'Normal\',\'Anormal\',\'No Registra\'])\n')
    f.write('            osteomuscular = st.selectbox("Osteomuscular", [\'Normal\',\'Anormal\',\'No Registra\'])\n')
    f.write('\n')
    f.write('    with tab3:\n')
    f.write('        st.markdown("#### Episodios de COVID-19 registrados en historia clínica")\n')
    f.write('        col_c1, col_c2, col_c3 = st.columns(3)\n')
    f.write('        with col_c1:\n')
    f.write('            covid1 = st.checkbox("Primer episodio")\n')
    f.write('        with col_c2:\n')
    f.write('            covid2 = st.checkbox("Segundo episodio")\n')
    f.write('        with col_c3:\n')
    f.write('            covid3 = st.checkbox("Tercer episodio")\n')
    f.write('        st.caption("Los antecedentes de COVID-19 se incluyen por su asociación documentada con secuelas funcionales.")\n')
    f.write('\n')
    f.write('with col_right:\n')
    f.write('    st.markdown("### 📊 Resumen del paciente")\n')
    f.write('    sistemas_alterados = sum([\n')
    f.write('        cardiovascular != \'Normal\', pulmonar != \'Normal\',\n')
    f.write('        neurologico != \'Normal\', mental != \'Normal\', osteomuscular != \'Normal\'\n')
    f.write('    ])\n')
    f.write('    covid_total = int(covid1) + int(covid2) + int(covid3)\n')
    f.write('\n')
    f.write('    st.info(f"""\n')
    f.write('**Edad:** {edad} años  \n')
    f.write('**IMC:** {imc_val} — {imc_cat}  \n')
    f.write('**Género:** {genero}  \n')
    f.write('**Sistemas alterados:** {sistemas_alterados} de 5  \n')
    f.write('**Disnea:** {escala_disnea.lower()}  \n')
    f.write('**COVID previo:** {covid_total} episodio(s)  \n')
    f.write('**Clasificación SUI:** {clasifisui[:30]}...\n')
    f.write('    """)\n')
    f.write('\n')
    f.write('    st.markdown("---")\n')
    f.write('    predict = st.button("⚡ ANALIZAR RIESGO", use_container_width=True, type="primary")\n')
    f.write('\n')
    f.write('# ── Predicción ────────────────────────────────────────────────────────────────\n')
    f.write('if predict:\n')
    f.write('    with st.spinner("Evaluando con Random Forest..."):\n')
    f.write('        time.sleep(0.4)\n')
    f.write('\n')
    f.write('        input_data = {\n')
    f.write('            \'EDAD\':           edad,\n')
    f.write('            \'PESO\':           peso,\n')
    f.write('            \'TALLA\':          talla,\n')
    f.write('            \'IMC\':            imc_val,\n')
    f.write('            \'CARDIOVASCULAR\': NORMAL_MAP[cardiovascular],\n')
    f.write('            \'PULMONAR\':       NORMAL_MAP[pulmonar],\n')
    f.write('            \'NEUROLÓGICO\':    NORMAL_MAP[neurologico],\n')
    f.write('            \'MENTAL\':         NORMAL_MAP[mental],\n')
    f.write('            \'OSTEOMUSCULAR\':  NORMAL_MAP[osteomuscular],\n')
    f.write('            \'RESULTADOIMC\':   IMC_MAP[resultado_imc],\n')
    f.write('            \'ESCALA DISNEA\':  DISNEA_MAP[escala_disnea],\n')
    f.write('            \'CLASIFISUI\':     CLASIFISUI_MAP[clasifisui],\n')
    f.write('            \'covid1\':         int(covid1),\n')
    f.write('            \'covid2\':         int(covid2),\n')
    f.write('            \'covid3\':         int(covid3),\n')
    f.write('            \'GENERO\':         GENERO_MAP[genero],\n')
    f.write('        }\n')
    f.write('\n')
    f.write('        X_input = pd.DataFrame([{k: input_data.get(k, 0) for k in feature_names}]).fillna(0)\n')
    f.write('        proba   = model.predict_proba(X_input)[0][1]\n')
    f.write('        pred    = model.predict(X_input)[0]\n')
    f.write('        pct     = int(proba * 100)\n')
    f.write('\n')
    f.write('        if proba >= 0.6:\n')
    f.write('            risk_level = "Alto";  risk_color = "#ef4444"; risk_icon = "🔴"; risk_class = "risk-high"\n')
    f.write('        elif proba >= 0.35:\n')
    f.write('            risk_level = "Medio"; risk_color = "#f59e0b"; risk_icon = "🟡"; risk_class = "risk-med"\n')
    f.write('        else:\n')
    f.write('            risk_level = "Bajo";  risk_color = "#10b981"; risk_icon = "🟢"; risk_class = "risk-low"\n')
    f.write('\n')
    f.write('        st.markdown(\'<div class="result-panel">\', unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('        col_r1, col_r2 = st.columns([1, 1])\n')
    f.write('        with col_r1:\n')
    f.write('            if pred == 1:\n')
    f.write('                st.markdown(f"### {risk_icon} Discapacidad predicha")\n')
    f.write('                st.error("El modelo indica presencia de discapacidad")\n')
    f.write('            else:\n')
    f.write('                st.markdown("### ✅ Sin discapacidad predicha")\n')
    f.write('                st.success("El modelo indica ausencia de discapacidad")\n')
    f.write('\n')
    f.write('            st.markdown(f\'<span class="risk-badge {risk_class}">{risk_icon} Riesgo {risk_level}</span>\', unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('        with col_r2:\n')
    f.write('            st.markdown(f"""\n')
    f.write('            <div style="text-align:center; padding: 1rem;">\n')
    f.write('                <div style="font-family:\'JetBrains Mono\',monospace; font-size:0.7rem; color:#64748b; text-transform:uppercase; letter-spacing:0.1em; margin-bottom:0.3rem;">Probabilidad de discapacidad</div>\n')
    f.write('                <div class="prob-percent">{pct}<span style="font-size:1.5rem; color:#64748b;">%</span></div>\n')
    f.write('                <div class="prob-bar-container" style="margin-top:0.5rem;">\n')
    f.write('                    <div style="height:12px; width:{pct}%; background:linear-gradient(90deg,{risk_color},#7c3aed); border-radius:40px;"></div>\n')
    f.write('                </div>\n')
    f.write('                <div style="display:flex; justify-content:space-between; font-size:0.65rem; color:#64748b; margin-top:0.3rem;">\n')
    f.write('                    <span>0%</span><span>50%</span><span>100%</span>\n')
    f.write('                </div>\n')
    f.write('            </div>\n')
    f.write('            """, unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('        with st.expander("📖 Explicabilidad del modelo — ¿por qué este resultado?"):\n')
    f.write('            st.markdown("""\n')
    f.write('**Features con mayor peso en el modelo (por importancia Random Forest):**\n')
    f.write('\n')
    f.write('| Feature | Importancia | Interpretación |\n')
    f.write('|---|---|---|\n')
    f.write('| EDAD | 17.8% | Mayor edad → mayor riesgo acumulado |\n')
    f.write('| IMC | 14.3% | Obesidad asociada a limitaciones funcionales |\n')
    f.write('| TALLA / PESO | 13.9% | Composición corporal relacionada con movilidad |\n')
    f.write('| CLASIFISUI | 9.7% | Clasificación de riesgo integral del paciente |\n')
    f.write('| MENTAL / RESULTADOIMC | 8.8% | Estado mental y nutricional como indicadores |\n')
    f.write('\n')
    f.write('*Modelo entrenado con 1,753 pacientes colombianos con enfermedades crónicas.*  \n')
    f.write('*Variables socioeconómicas excluidas intencionalmente para evitar sesgo algorítmico.*\n')
    f.write('            """)\n')
    f.write('\n')
    f.write('        st.markdown("""\n')
    f.write('        <div style="font-family:\'JetBrains Mono\',monospace; font-size:0.65rem; color:#475569; margin-top:1.2rem; padding-top:1rem; border-top:1px solid #1e293b;">\n')
    f.write('        ⚠ Este resultado es una estimación estadística del modelo y no reemplaza la evaluación clínica profesional.\n')
    f.write('        &nbsp;·&nbsp; Random Forest · GridSearchCV · 360 combinaciones · 5-fold CV\n')
    f.write('        </div>\n')
    f.write('        """, unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('        st.markdown(\'</div>\', unsafe_allow_html=True)\n')
    f.write('\n')
    f.write('# ── Footer ────────────────────────────────────────────────────────────────────\n')
    f.write('st.markdown("""\n')
    f.write('<div class="footer">\n')
    f.write('    PRÁCTICA 4 — ANALÍTICA DE DATOS 2025 &nbsp;·&nbsp; UNIVERSIDAD PONTIFICIA BOLIVARIANA &nbsp;·&nbsp; MEDELLÍN, COLOMBIA<br>\n')
    f.write('    Variables socioeconómicas excluidas del modelo para evitar sesgo algorítmico · Uso exclusivamente académico\n')
    f.write('</div>\n')
    f.write('""", unsafe_allow_html=True)\n')
    f.write('\n')
print('app.py generado correctamente')
print('Ejecutar: python -m streamlit run app.py')
````
